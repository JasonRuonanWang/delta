```python
data = np.random.uniform(0.0, 1.0, 1000)
```

```python
data_ft = dask_client.scatter(data, broadcast=True)
```

```python
func_dict = {"sum": lambda x: x.sum(), "mean": lambda x: x.mean()}
ft_funs = dask_client.scatter(func_dict, broadcast = True)
```

```python
def f(data_ft, ft_funs, what=0):
    #import numpy as np
    
    print("Hallohallohallo")
    
    if what == 0:
        return ft_funs["sum"](data_ft)
    
    elif what == 1:
        return ft_funs["mean"](data_ft)
    
    else:
        return(-1.0)
```

```python
res = dask_client.submit(f, data_ft, ft_funs, 0)
```
